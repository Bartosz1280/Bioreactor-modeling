# Different growth models implemented so far

```python
import numpy as np
from gekko import GEKKO
import matplotlib.pyplot as plt
import math


from Bioreactor import SimpleBioreactor, SimulationInput
```

```python
sim_input_monod_subtrate = SimulationInput(mu_max = 0.86, K_s = 0.0138, 
                                 Y_xs = 1.28, K_d = 3.3e-3, 
                                 X = 0.005, S = 0.07,
                                    # By default growth_type is set to monod_substrate
                                    )
time_arr = np.linspace(0,15,1000)

br1 = SimpleBioreactor(time_arr, 
                       sim_input_monod_subtrate.get_inputs(),
                       name="Monod without biomass")
br1.solve()
```

```python
ax1 = br1.plot()
```

```python
sim_input_monod_subtrate = SimulationInput(mu_max = 0.86, K_s = 0.0138, 
                                 Y_xs = 1.28, K_d = 3.3e-3, 
                                 X = 0.005, S = 0.07,
                                growth_type = "blackman"
                                    )
time_arr = np.linspace(0,15,1000)

br2 = SimpleBioreactor(time_arr,
                       sim_input_monod_subtrate.get_inputs(),
                       name="Blackman")
br2.solve()
```

```python
ax2 = br2.plot()
```

```python
sim_input_monod_subtrate = SimulationInput(mu_max = 0.86, K_s = 0.0138, 
                                 Y_xs = 1.28, K_d = 3.3e-3, 
                                 X = 0.005, S = 0.07,
                                growth_type = "contois"
                                    )
time_arr = np.linspace(0,7,1000)

br3 = SimpleBioreactor(time_arr,
                       sim_input_monod_subtrate.get_inputs(),
                       name="Contois model")
br3.solve()

```
